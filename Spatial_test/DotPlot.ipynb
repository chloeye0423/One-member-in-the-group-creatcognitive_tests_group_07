```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import ttest_ind

tests = ['ANS','Memory','Spatial Mapping','Maths']
mean_values_M = [81.9083, 37.7778,94.2486,70.19]
mean_values_F = [81.7089,62.6667,85.8929,72.0333]
p_values = [0.9851,0.0501,0.0469,0.6783]
std_error_M = [7.18779,6.62539,2.5669,3.21293]
std_error_F = [6.8517,7.97032,2.3229,2.60254]
fig, ax = plt.subplots()

ax.plot(range(len(tests)), mean_values_M, 'o', color = 'blue')
ax.plot(range(len(tests)), mean_values_F, 'o', color = 'red')
ax.errorbar(range(len(tests)),mean_values_M,yerr=std_error_M,fmt='o',color='blue')
ax.errorbar(range(len(tests)),mean_values_F,yerr=std_error_F,fmt='o',color='red')

for i, (mean, p_value) in enumerate(zip(mean_values_M, p_values)):
    ax.text(i,mean,f'{mean:.2f}',ha='right',va='top')
    ax.text(i,mean,f'p-value: {p_value: .4f}', ha = 'left', va='top')
    
for i, (mean, p_value) in enumerate(zip(mean_values_F, p_values)):
    ax.text(i,mean,f'{mean:.2f}',ha='right',va='bottom')
    ax.text(i,mean,f'p-value: {p_value: .4f}', ha = 'left', va='top')
    
ax.set_xlabel('Tests')
ax.set_ylabel('Mean value')
ax.set_title('Dot plot for t-tests for the cognitive tasks')

ax.set_xticks(range(len(tests)))
ax.set_xticklabels(tests)

plt.show()


#Figure: Dot plot for results of statistical analysis from each cognitive test. Male values are denoted in blue and female values denoted in red. 
```
