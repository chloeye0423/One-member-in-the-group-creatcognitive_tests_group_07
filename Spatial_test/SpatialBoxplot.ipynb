```python
import matplotlib.pyplot as plt

gender = ['Male','Female','Male','Female','Male','Female','Female','Male','Male','Female','Male','Female','Female','Female','Female','Female','Female','Male','Female','Female','Female']
percentscore = [79.95,80,95,92.5,100,77.5,87.5,95,90,82.5,100,92.5,87.5,82.5,95,80,95,100,65,85,100]

data = {'Male':[], 'Female':[]}

for g, ps in zip(gender, percentscore):
    data[g].append(ps)

score_values = [data['Male'],data['Female']]

plt.boxplot(score_values,
            labels=['Male','Female'])
plt.xlabel('Gender')
plt.ylabel('Score (%)')
plt.title('Relationship Between Gender and Score in the Spatial Reasoning Task')

plt.show()
```
