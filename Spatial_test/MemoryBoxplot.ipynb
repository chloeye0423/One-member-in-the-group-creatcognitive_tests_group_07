```python
import matplotlib.pyplot as plt

gender = ['Male','Female','Male','Male','Female','Male','Female','Male','Female','Male','Female','Male','Female','Female','Female','Female','Female','Female','Female','Male','Female','Female','Male','Female']
percentscore = [60,40,0,20,20,40,100,60,80,40,60,60,100,40,100,80,0,60,40,40,40,80,20,100]

data = {'Male':[], 'Female':[]}

for g, ps in zip(gender, percentscore):
    data[g].append(ps)

score_values = [data['Male'],data['Female']]

plt.boxplot(score_values,
            labels=['Male','Female'])
plt.xlabel('Gender')
plt.ylabel('Score (%)')
plt.title('Relationship Between Gender and Score in the Memory Test')

plt.show()
```
