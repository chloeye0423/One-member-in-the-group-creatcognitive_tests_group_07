```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming your CSV file is named 'your_data.csv' and has columns 'Gender' and 'Score in percentage'
df = pd.read_csv('ANS_results.csv')

# Extracting data from the DataFrame
gender = df['Gender'].tolist()
percentscore = df['Score in percentage'].tolist()

data = {unique_gender: [] for unique_gender in set(gender)}

for g, ps in zip(gender, percentscore):
    data[g].append(ps)

score_values = list(data.values())
labels = list(data.keys())

plt.boxplot(score_values, labels=labels)
plt.xlabel('Gender')
plt.ylabel('Score (%)')
plt.title('Relationship Between Gender and Score in the ANS Task')

plt.show()
```
